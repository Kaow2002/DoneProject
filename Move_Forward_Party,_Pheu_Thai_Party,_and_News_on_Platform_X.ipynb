```python
from ntscraper import Nitter
import pandas as pd
```

```python
scraper = Nitter(0)
```

```python
def get_tweets(name, modes):
    tweets = scraper.get_tweets(name, mode = modes,since='2023-02-14',until='2023-05-14')
    final_tweets = []
    for x in tweets['tweets']:
        data = [x['link'], x['text'],x['date'],x['stats']['likes'],x['stats']['comments'],x['stats']['retweets']]
        final_tweets.append(data)
    dat= pd.DataFrame(final_tweets, columns =['twitter_link','text','date','likes','comments','retweets'])
    return dat
```

```python
import re
from pythainlp import word_tokenize
from pythainlp.corpus import thai_stopwords
```

```python
def text_process_thai(text):
    #ลบสัญลักษณ์และemoji
    text = re.sub(r'[^\w\sก-๙]', '', text)

    tokens = word_tokenize(text, engine='newmm')

    #ลบstopword
    tokens = [word for word in tokens if word not in thai_stopwords()]
    tokens = [word for word in tokens if word not in ['ผลิตโดย'] and not word.isdigit()]

    final_text = " ".join(tokens)
    final_text = re.sub(r'\d+', '', final_text)
    return final_text

pita = pita.astype(str)
ing = ing.astype(str)
sedtha = sedtha.astype(str)
voice = voice.astype(str)
one = one.astype(str)
morningch3 = morningch3.astype(str)
thairath = thairath.astype(str)
kaoklai = kaoklai.astype(str)
pheuthai = pheuthai.astype(str)


pita['text.tokens'] = pita['text'].apply(text_process_thai)
ing['text.tokens'] = ing['text'].apply(text_process_thai)
sedtha['text.tokens'] = sedtha['text'].apply(text_process_thai)
one['text.tokens'] = one['text'].apply(text_process_thai)
morningch3['text.tokens'] = morningch3['text'].apply(text_process_thai)
thairath['text.tokens'] = thairath['text'].apply(text_process_thai)
voice['text.tokens'] = voice['text'].apply(text_process_thai)
kaoklai['text.tokens'] = kaoklai['text'].apply(text_process_thai)
pheuthai['text.tokens'] = pheuthai['text'].apply(text_process_thai)

```

```python
#ลบ row ที่เป็น list ว่างออก
pita = pita[pita['text.tokens'].apply(lambda x: x not in ['nan', ''])]
ing = ing[ing['text.tokens'].apply(lambda x: x not in ['nan', ''])]
sedtha = sedtha[sedtha['text.tokens'].apply(lambda x: x not in ['nan', ''])]
one = one[one['text.tokens'].apply(lambda x: x not in ['nan', ''])]
thairath = thairath[thairath['text.tokens'].apply(lambda x: x not in ['nan', ''])]
voice = voice[voice['text.tokens'].apply(lambda x: x not in ['nan', ''])]
morningch3 = morningch3[morningch3['text.tokens'].apply(lambda x: x not in ['nan', ''])]
kaoklai = kaoklai[kaoklai['text.tokens'].apply(lambda x: x not in ['nan', ''])]
pheuthai = pheuthai[pheuthai['text.tokens'].apply(lambda x: x not in ['nan', ''])]
```
